## *Opdracht 6*: Meten en fitten, Leerdoel

```python
#import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit 
# Zet hier jouw data in een voor jou handig formaat

m_sensor1 = np.array([108,104,107,106,106])
m_sensor2 = np.array([243,243,241,238,238])
m_sensor3 = np.array([378,377,377,378,378])
m_sensor4 = np.array([480,484,484,486,485])
m_sensor5 = np.array([558,559,560,554,558])
m_sensor7 = np.array([8,8,6,7,12])
m_sensor8 = np.array([146,143,134,136,137])
m_sensor9 = np.array([427,426,419,427,429])
m_sensor10 = np.array([481,487,484,489,469])

m_sensor = np.array([ np.mean(m_sensor1),np.mean(m_sensor2),np.mean(m_sensor3),np.mean(m_sensor4),np.mean(m_sensor5),np.mean(m_sensor7),np.mean(m_sensor8),np.mean(m_sensor9),np.mean(m_sensor10)])
std_sensor = np.array([ np.std(m_sensor1),np.std(m_sensor2),np.std(m_sensor3),np.std(m_sensor4),np.std(m_sensor5),np.std(m_sensor7),np.std(m_sensor8),np.std(m_sensor9),np.std(m_sensor10)])
u_sensor = std_sensor / np.sqrt(5)

#576.4
m_gewogen = np.array([ 103.7, 176.2, 267.8, 319.4, 389.4, 72.5, 161.2, 231.2, 281.8 ])
u_gewogen = 0.1
```

```python
# Plot hier de meting en fit een lijn door de metingen.
# Maak gebruik van wat je geleerd hebt bij het practicum.

def fitfunctie(x,a):
    return a*x 

popt, pcov = curve_fit(fitfunctie, m_gewogen, m_sensor)


x = np.linspace(0,500,1000)
y = fitfunctie(x, *popt)

plt.figure()
plt.title("ijkplot van de kracht sensor")
plt.errorbar(m_gewogen, m_sensor, u_gewogen, u_sensor, "o", label="data")
plt.plot(x,y,"r--", label="fitfunctie")
plt.xlabel('daadwerkelijk gewicht (g)')
plt.ylabel('meetwaardes van de sensor (bits)')
plt.xlim(0,500)
plt.ylim(0,700) 
plt.legend()
plt.show()

```
